```python


# 파일 경로에 맞게 수정하세요
test_with_predictions_path = "../denoised_results.csv" # 텍스트가 바뀐 후 파일 입력

# 파일 불러오기
test_with_predictions = pd.read_csv(test_with_predictions_path)

# text 칼럼의 길이를 계산하고, 결측치는 길이 0으로 계산
test_with_predictions['text_length'] = test_with_predictions['denoised_text'].apply(lambda x: len(x) if isinstance(x, str) else 0)

# Sort the DataFrame by the new column in ascending order
test_with_predictions_sorted = test_with_predictions.sort_values(by='text_length')

# 결과 확인
test_with_predictions_sorted

```

```python
# 조건에 맞는 행(길이가 8 이하이거나 텍스트가 'missing title')을 필터링하여 저장
remaining_rows = test_with_predictions[(test_with_predictions['text_length'] <= 8) | (test_with_predictions['text'] == 'missing title')]

# 나머지 행은 별도의 DataFrame에 저장
filtered_rows = test_with_predictions.drop(remaining_rows.index).reset_index(drop=True)

print(len(remaining_rows))

remaining_rows
```

```python
# test_df의 text 열 삭제
filtered_rows = filtered_rows.drop(columns=["text"])
remaining_rows = remaining_rows.drop(columns=["text"])

# denoised_text 열 이름을 text로 변경
filtered_rows = filtered_rows.rename(columns={"denoised_text": "text"})
remaining_rows = remaining_rows.rename(columns={"denoised_text": "text"})

filtered_rows
```

```python
filtered_rows = filtered_rows.drop(columns=["is_noise"])
remaining_rows = remaining_rows.drop(columns=["is_noise"])

filtered_rows = filtered_rows.drop(columns=["text_length"])
remaining_rows = remaining_rows.drop(columns=["text_length"])

```

```python
# 인덱스 재정렬 후 각각의 파일로 저장 (인덱스 없이)
filtered_file_path = "../denoised_results_wo_error.csv"
filtered_rows.reset_index(drop=True, inplace=True)
filtered_rows.to_csv(filtered_file_path, index=False)

# 인덱스 재정렬 후 각각의 파일로 저장 (인덱스 없이)
remaining_rows_file_path = "../denoised_results_error.csv"
remaining_rows.reset_index(drop=True, inplace=True)
remaining_rows.to_csv(remaining_rows_file_path, index=False)
```
